### Feature Engineering

#### Missing Values

```python
pd.pandas.set_option('display.max_columns', None)
test_dataset=pd.read_csv('test.csv')
test_dataset.head()
```

```python
dataset.head()
```

```python
## Let us capture all the nan values
## First lets handle Categorical features which are missing
features_nan=[feature for feature in test_dataset.columns if test_dataset[feature].isnull().sum()>1 and test_dataset[feature].dtypes=='O']

for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(test_dataset[feature].isnull().mean(),4)))
```

```python
## Replace missing value with a new label
def replace_cat_feature(test_dataset,features_nan):
    data=test_dataset.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data

test_dataset=replace_cat_feature(test_dataset,features_nan)

test_dataset[features_nan].isnull().sum()
```

```python
## Now lets check for numerical variables the contains missing values
numerical_with_nan=[feature for feature in test_dataset.columns if test_dataset[feature].isnull().sum()>1 and test_dataset[feature].dtypes!='O']

## We will print the numerical nan variables and percentage of missing values

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(test_dataset[feature].isnull().mean(),4)))
```

```python
## Replacing the numerical Missing Values

for feature in numerical_with_nan:
    ## We will replace by using median since there are outliers
    median_value=test_dataset[feature].median()
        
    ## create a new feature to capture nan values
    test_dataset[feature+'nan']=np.where(test_dataset[feature].isnull(),1,0)
    test_dataset[feature].fillna(median_value,inplace=True)
    
test_dataset[numerical_with_nan].isnull().sum()    
```

```python
test_dataset.head(10)
```

```python
## Temporal Variables (Date Time Variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:      
    test_dataset[feature]=test_dataset['YrSold']-test_dataset[feature]
```

```python
test_dataset[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

#### Numerical Variables

Since the numerical variables are skewed we will perform log normal distribution

```python
import numpy as np
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea']

for feature in num_features:
    test_dataset[feature]=np.log(test_dataset[feature])
```

```python
test_dataset.head()
```

#### Handling Rare Categorical Feature

We will remove categorical variables that are present less than 1% of the observations

```python
categorical_features=[feature for feature in test_dataset.columns if test_dataset[feature].dtype=='O']
```

```python
categorical_features
```

```python
test_dataset.head(100)
```

```python
for feature in categorical_features:
    labels_ordered=test_dataset.groupby([feature]).mean().sort_values('Id').index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    test_dataset[feature]=test_dataset[feature].map(labels_ordered)
```
